## **<span style="color:cyan">Part a.</span>**

```python
Repeat_Num = 100
List_Len = 1000
List = list(np.random.randint(0, 10000, List_Len))
```

```python
BubbleSort_dist = np.zeros(Repeat_Num)
for i in range(Repeat_Num):
    rd.shuffle(List)
    start = time.time()
    bubble_sort(List)
    end = time.time()
    BubbleSort_dist[i] = end - start
```

```python
QuickSort_dist = np.zeros(Repeat_Num)
for i in range(Repeat_Num):
    rd.shuffle(List)
    start = time.time()
    quick_sort(List)
    end = time.time()
    QuickSort_dist[i] = end - start
```

```python
InsertionSort_dist = np.zeros(Repeat_Num)
for i in range(Repeat_Num):
    rd.shuffle(List)
    start = time.time()
    insertion_sort(List)
    end = time.time()
    InsertionSort_dist[i] = end - start
```

```python
MergeSort_dist = np.zeros(Repeat_Num)
for i in range(Repeat_Num):
    rd.shuffle(List)
    start = time.time()
    merge_sort(List)
    end = time.time()
    MergeSort_dist[i] = end - start
```

```python
def salt_pepper(image, ratio):
    noisy_image = image.copy()
    total_pixels = image.size
    num_noise_pixels = int(ratio * total_pixels)
    noisy_indices = np.random.choice(total_pixels, num_noise_pixels, replace = False)
    noisy_image.flat[noisy_indices[0:num_noise_pixels//2]] = 255
    noisy_image.flat[noisy_indices[num_noise_pixels//2:-1]] = 0
    return noisy_image

def gaussian(image, ratio):
    noisy_image = image.copy()
    total_pixels = image.size
    num_noise_pixels = int(ratio * total_pixels)
    mean_x, mean_y = np.random.randint(300, 700, 2) 
    std_x, std_y = np.random.randint(100, 600, 2) 
    noisy_indices_x = np.round(np.random.normal(500, 400, [int(5 * num_noise_pixels), 1]))
    noisy_indices_y = np.round(np.random.normal(500, 400, [int(5 * num_noise_pixels), 1]))
    noisy_indices_x = (noisy_indices_x[(noisy_indices_x < 1000) & (noisy_indices_x > 0)])[:num_noise_pixels]
    noisy_indices_y = (noisy_indices_y[(noisy_indices_y < 1000) & (noisy_indices_y > 0)])[:num_noise_pixels]
    noisy_indices_x = np.reshape([int(noisy_indices_x[i]) for i in range(len(noisy_indices_x))], [num_noise_pixels, 1])
    noisy_indices_y = np.reshape([int(noisy_indices_y[i]) for i in range(len(noisy_indices_y))], [num_noise_pixels, 1])
    noisy_indices = np.append(noisy_indices_x, noisy_indices_y, axis = 1)
    noisy_indices = [tuple(noisy_indices[i]) for i in range(len(noisy_indices))]
    for i in range(len(noisy_indices)//2):
        noisy_image[tuple(noisy_indices[i])] = 0
        noisy_image[tuple(noisy_indices[i + len(noisy_indices)//2])] = 255
    return noisy_image

def exponential(image, ratio):
    noisy_image = image.copy()
    total_pixels = image.size
    num_noise_pixels = int(ratio * total_pixels)
    noisy_indices_x = np.random.exponential(1, [int(2 * num_noise_pixels), 1])
    noisy_indices_y = np.random.exponential(1, [int(2 * num_noise_pixels), 1])
    noisy_indices_x = np.round(len(image) / max(noisy_indices_x) * 5 * noisy_indices_x)
    noisy_indices_y = np.round(len(image) / max(noisy_indices_y) * 5 * noisy_indices_y)
    noisy_indices_x = (noisy_indices_x[(noisy_indices_x < 1000) & (noisy_indices_x > 0)])[:num_noise_pixels]
    noisy_indices_y = (noisy_indices_y[(noisy_indices_y < 1000) & (noisy_indices_y > 0)])[:num_noise_pixels]
    noisy_indices_x = np.reshape([int(noisy_indices_x[i]) for i in range(len(noisy_indices_x))], [num_noise_pixels, 1])
    noisy_indices_y = np.reshape([int(noisy_indices_y[i]) for i in range(len(noisy_indices_y))], [num_noise_pixels, 1])
    noisy_indices = np.append(noisy_indices_x, noisy_indices_y, axis = 1)
    noisy_indices = [tuple(noisy_indices[i]) for i in range(len(noisy_indices))]
    for i in range(len(noisy_indices)//2):
        noisy_image[tuple(noisy_indices[i])] = 0
        noisy_image[tuple(noisy_indices[i + len(noisy_indices)//2])] = 255
    return noisy_image
```

```python
ratios = [0.1, 0.15, 0.2, 0.3, 0.4, 0.5]
image = cv2.imread('cookie-monster.jpg', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (20, 10))
for i in range(len(ratios)):
    plt.subplot(3, 6, i+1)
    noisy_image = salt_pepper(image, ratios[i])
    plt.imshow(noisy_image, cmap='gray')
    plt.axis = 'Off'
    plt.title('SP Noise: {}%'.format(ratios[i]  * 100))

    plt.subplot(3, 6, i+7)
    noisy_image = gaussian(image, ratios[i])
    plt.imshow(noisy_image, cmap='gray')
    plt.axis = 'Off'
    plt.title('Gaussian Noise: {}%'.format(ratios[i]  * 100))

    plt.subplot(3, 6, i+13)
    noisy_image = exponential(image, ratios[i])
    plt.imshow(noisy_image, cmap='gray')
    plt.axis = 'Off'
    plt.title('Exponential Noise: {}%'.format(ratios[i]  * 100))
```
